```python
import ipywidgets as widgets
from IPython.display import display, clear_output

data = {"so_du": 50000, "user": "nghiadao_ueh"}
title = widgets.HTML("<h2>🚌 BUSPAY - CHẠM LÀ ĐI</h2>")
input_km = widgets.IntText(value=5, description='Quãng đường(km):', style={'description_width': 'initial'})
type_ticket = widgets.Dropdown(options=[('Sinh viên (3k)', 3000), ('Vé thường (7k)', 7000)], description='Loại vé:')
btn_pay = widgets.Button(description='QUÉT MÃ THANH TOÁN', button_style='success', layout=widgets.Layout(width='200px'))
output = widgets.Output()
def on_pay_clicked(b):
    with output:
        clear_output()
        phi = type_ticket.value
        if input_km.value > 15:
            phi += (input_km.value - 15) * 500
        if data["so_du"] >= phi:
            data["so_du"] -= phi
            print(f"✅ THANH TOÁN THÀNH CÔNG!")
            print(f"➖ Đã trừ: {phi:,} VNĐ")
            print(f"💰 Số dư ví còn lại: {data['so_du']:,} VNĐ")
            print("-" * 30)
        else:
            print("❌ LỖI: Số dư trong ví không đủ!")
btn_pay.on_click(on_pay_clicked)
display(title, type_ticket, input_km, btn_pay, output)
```
